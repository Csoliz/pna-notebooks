ANTS is picky about the affine information in the image.  The subject image doesn't have a proper affine:

```python
anat_hdr = anat_img.get_header()
print(anat_hdr)
```

For ANTS to work properly, we will write a new copy with a proper affine.  In fact nibabel will guess a good-enough affine for us; we use that.

```python
affine = anat_img.get_affine()
affine
```

```python
fixed_anat = nib.Nifti1Image(anat_img.get_data(), affine, anat_hdr)
```

```python
# Save in the current directory
nib.save(fixed_anat, 'fixed_anat.nii')
```
