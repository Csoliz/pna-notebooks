```python
import numpy as np
import matplotlib.pyplot as plt
```

The MNI templates are here: <http://www.bic.mni.mcgill.ca/~vfonov/icbm/2009/mni_icbm152_nlin_asym_09a_nifti.zip>

```python
import os
import nibabel as nib
```

```python
MNI_PATH = 'mni_icbm152_nlin_asym_09a'
```

```python
gm_fname = os.path.join(MNI_PATH, 'mni_icbm152_gm_tal_nlin_asym_09a.nii')
wm_fname = os.path.join(MNI_PATH, 'mni_icbm152_wm_tal_nlin_asym_09a.nii')
csf_fname = os.path.join(MNI_PATH, 'mni_icbm152_csf_tal_nlin_asym_09a.nii')
```

```python
gm_img = nib.load(gm_fname)
gm_data = gm_img.get_data()
wm_img = nib.load(wm_fname)
wm_data = wm_img.get_data()
csf_img = nib.load(csf_fname)
csf_data = csf_img.get_data()
csf_data.shape
```

```python
mid_gm = gm_data[:, :, 94]
mid_wm = wm_data[:, :, 94]
mid_csf = csf_data[:, :, 94]
```

```python
plt.imshow(np.hstack((mid_gm, mid_wm, mid_csf)), cmap="gray")
```

```python
t2_like = mid_gm * 0.5 + mid_wm * 0.2 + mid_csf * 0.8
t1_like = mid_gm * 0.7 + mid_wm * 1.0 + mid_csf * 0.1
plt.imshow(np.hstack((t2_like, t1_like)), cmap="gray")
```

```python
np.savez('t1_t2_like.npz', {'t2_like': t2_like, 't1_like': t1_like})
```

```python
plt.plot(t2_like.ravel(), t1_like.ravel(), '.')
```

```python
hgram_registered, x_edges, y_edges = np.histogram2d(t2_like.ravel(), t1_like.ravel(), 100)
hgram_registered.max()
```

```python
plt.imshow(np.flipud(np.clip(hgram_registered, 0, 20)), cmap='gray', interpolation='nearest')
```

Move t1 like image 30 pixels down

```python
t1_like_moved = np.zeros_like(t1_like)
t1_like_moved[20:, :] = t1_like[:-20, :]
plt.imshow(t1_like_moved, cmap="gray")
```

```python
plt.plot(t2_like.ravel(), t1_like_moved.ravel(), '.')
```

```python
hgram_off, x_edges, y_edges = np.histogram2d(t2_like.ravel(), t1_like_moved.ravel(), 100)
plt.imshow(np.flipud(np.clip(hgram_off, 0, 20)), cmap='gray', interpolation='nearest')
```
